```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X + 2*Y
X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X1, Y1 = np.meshgrid(X1, Y1)
Z1 = X1 + 2*Y1 + 20
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X1, Y1, Z1)
show()
```
